```python
import pandas as pd
df=pd.read_csv("Walmart_Sales.csv")
# 1
df.head(5)
```

```python
print(df.dtypes)
print("\nCount of missing values per column:")
print(df.isnull().sum())
```

```python
# Function to calculate relevant statistics
def calculate_statistics(df):
    stats = {}
    
    # Numerical columns
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numerical_cols:
        stats[col] = {
            'count': df[col].count(),
            'mean': df[col].mean(),
            'median': df[col].median(),
            'std_dev': df[col].std(),
            'min': df[col].min(),
            'max': df[col].max(),
            '25%': df[col].quantile(0.25),
            '50%': df[col].quantile(0.50),
            '75%': df[col].quantile(0.75)
        }
    
    # Categorical columns
    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        stats[col] = {
            'count': df[col].count(),
            'unique': df[col].nunique(),
            'top': df[col].mode()[0] if not df[col].mode().empty else None,
            'freq': df[col].value_counts().iloc[0] if not df[col].value_counts().empty else None
        }
    
    # Date columns
    date_cols = df.select_dtypes(include=['datetime64']).columns
    for col in date_cols:
        stats[col] = {
            'count': df[col].count(),
            'min': df[col].min(),
            'max': df[col].max(),
            'range': df[col].max() - df[col].min()
        }
    
    return stats

# Calculate statistics
statistics = calculate_statistics(df)

# Print the statistics
for column, stat in statistics.items():
    print(f"Statistics for '{column}':")
    for key, value in stat.items():
        print(f"  {key}: {value}")
    print()
```

Using transform with a Custom Function
Define a function scaled_sales(s) which z-score scales Weekly_Sales per Store. That is, for each store:  

$$ \text{z-score} = \frac{(\text{Weekly\_Sales} - \mu)}{\sigma} $$

where $$\mu$$ is the mean Weekly_Sales for that store, and $$\sigma$$ is the standard deviation.
1.	Use transform on the Weekly_Sales column grouped by Store to apply this function.
2.	Save it to a new column named ZScore_Sales.

## Question 10 - 10 points

Apply and GroupBy for Windowed Custom Calculation
Create a custom function that calculates the rolling coefficient of variation (CoV) for the last 4 weeks of Weekly_Sales per Store. The coefficient of variation is defined as:

$$\text{CoV} = \frac{\sigma}{\mu}$$

where $$\sigma$$ is the standard deviation and $$\mu$$ is the mean of the last 4 weeks of Weekly_Sales.
1.	Sort the DataFrame by Store and Date.
2.	Group by Store and use apply to apply a function that calculates the 4-week rolling CoV of Weekly_Sales.
3.	Create a new column Rolling_4W_CoV to store this value.

```python
df.sort_values(by=['Store', 'Date'], inplace=True)

df['Rolling_4W_CoV'] = df.groupby('Store')['Weekly_Sales'].transform(
    lambda x: x.rolling(window=4).std() / x.rolling(window=4).mean()
)

df.head(5)
```
