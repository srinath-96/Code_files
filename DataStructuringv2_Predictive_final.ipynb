```python
Flavor_m = sm.OLS.from_formula('Unit_Sales~ C(Brand)+Flavor+Price_per_Unit_Any_Merch+Price_per_Unit_No_Merch+Price_per_Unit_Price_Reductions_Only+Flavor:C(Brand):Price_per_Unit_Any_Merch+Flavor:C(Brand):Price_per_Unit_No_Merch+Flavor:C(Brand):Price_per_Unit_Price_Reductions_Only', x).fit()

# Print the summary of the regression
print(Flavor_m.summary())
```
